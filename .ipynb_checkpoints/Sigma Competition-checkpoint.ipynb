```python
from pandas  import *
from numpy  import *
options.display.max_columns = 100

data = pandas.io.parsers.read_csv("/Users/Eric/Documents/Data Analysis/fares.csv",quotechar='"', header=2)
data
```

```python
data.columns
    
```

```python
import geopy
import time
from geopy.geocoders import Nominatim
from geopy.exc import GeocoderTimedOut

geolocator = Nominatim()

def findLocation(x):
    try:
        location = geolocator.geocode(x['STATION']+" NYC")
        if location:
            x['LAT'] = location.latitude
            x['LON'] = location.longitude
        return x
    except GeocoderTimedOut as e:
        return x
data=data.apply(findLocation, axis=1)
data
```

```python
cleanedData = data[[col for col in ["LAT", "LON", "FF"]]]
cleanedData = cleanedData[notnull(cleanedData["LAT"])]
cleanedData
```

```python
cleanedData.to_csv("subways.csv")
```
