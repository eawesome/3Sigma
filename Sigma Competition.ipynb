```python
import geopy
import time
from geopy.geocoders import Nominatim
geolocator = Nominatim()

def findAddress(x):
    location = geolocator.geocode(x['STATION']+" NYC")
    time.sleep(.1)
    if location:
        x['ADDRESS'] = location.address
    return x

data=data.apply(findAddress, axis=1)
```
